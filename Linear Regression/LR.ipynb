```python
import numpy as np 

X = np.array([[1, 50], [1, 60], [1, 70], [1, 100]])
X
```

```python
y  = np.array([[10],[30],[40],[50]])
y
```

```python
X_T_X = (X.T).dot(X)
X_T_X
```

```python
from numpy.linalg import inv

X_T_X_inv = inv(X_T_X)

X_T_X_inv
```

```python
w = X_T_X_inv.dot(X.T).dot(y)
print(w)
```

```python
from matplotlib import pyplot as plt
import numpy as np 

%matplotlib inline

margin=10
X_min = X[:,1].max() - margin
X_max = X[:,1].max() + margin

X_sup = np.linspace(X_min, X_max, num=100)
y_model = w[0][0] + w[1][0]*X_sup

plt.xlim(X_min, X_max)
plt.ylim(0, y[:,0].max() + margin)

plt.scatter(X[:, 1], y[:,0],40,'g', 'o', alpha=0.8)

plt.plot(X_sup, y_model)

plt.show()
```

```python
#Pretty print python's float
def pretty_float(a, format_string='{0:.2f}'):
    return [format_string.format(v,i) for i, v in enumerate(a)]
```

```python
from sklearn.datasets import fetch_california_housing


cali_dataset = fetch_california_housing()

features = cali_dataset.data 
y = cali_dataset.target 

print(cali_dataset.DESCR)
print('\nMatrix Objects X Features (dimension): %s %s' % features.shape)
print('\nTarget variable y (dimension): %s' % y.shape)
```

```python
w = inv((features.T).dot(features)).dot(features.T).dot(y)
pretty_float(w)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(features, y)
```

```python
pretty_float(reg.coef_)
```

```python
y_pred = reg.predict(features)
y_true = y
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, \
                            r2_score

mean_squared_error(y_true, y_pred)
```

```python
mean_absolute_error(y_true, y_pred)
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


houses =  pd.read_csv('houses.csv')
houses.head()


figure = plt.figure(figsize=(10, 10))
markers = {'basic': 'o',
            'medium': 'v',
            'luxury': '*'}

for d1, d2, l in zip(houses.dim_1, houses.dim_2, houses.level):
    plt.scatter(d1, d2, s=200, marker=markers[l])

plt.show() 
```

```python
from sklearn.cluster import KMeans

X = houses[['dim_1', 'dim_2']]
```

```python
clust = KMeans(n_clusters=3).fit(X)
```

```python
[c1, c2, c3] = clust.cluster_centers_
print(clust.cluster_centers_)
```

```python
figure = plt.figure(figsize=(10, 10))
markers = {'basic': 'o',
            'medium': 'v',
            'luxury': '*'}

for d1, d2, l in zip(houses.dim_1, houses.dim_2, houses.level):
    plt.scatter(d1, d2, s=200, marker=markers[l])

plt.scatter(c1[0], c1[1], s=250, marker='x', c='black')
plt.scatter(c1[0], c1[1], s=250 * 5e2, c='black', alpha=0.1)

plt.scatter(c2[0], c2[1], s=250, marker='x', c='black')
plt.scatter(c2[0], c2[1], s=250 * 2e3, c='black', alpha=0.1)

plt.scatter(c3[0], c3[1], s=250, marker='x', c='black')
plt.scatter(c3[0], c3[1], s=250 * 2.5e2, c='black', alpha=0.1)

plt.show()
```

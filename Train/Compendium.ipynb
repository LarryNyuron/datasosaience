Типы маштнного обучения:
1. Классическое обучение (есть полный набор даных прихнаков, есть ли досмтуп к целевому или нет)
    1.1. Обучение с учителем - данные размечены - (экземпляр и что должно получиться):
      - задача регрессии - должны спрогнозировать непрерывнеые величины (пример - прогноз погоды )
      
      - и задачи классификации - (кто изображен на фото: фрукт)
      
      - кластеризация - (взяли и по кучкам разделили)
    1.2. Обучение без учителя - данные неразмечены - (есть данные, но нет идеального выхода) - нужно выявить закономерности или аномалии
      

2. Обучение с подкреплением - (задачи минимазации ошибки, когда нет напрямую взаимоджействия с данными, но можем считывать что-то косвенное, к примеру любая игра автоматизированная Флаппи бёрд) - система вознаграждений и штрафл.

3. Ансамбли алгоримтмов - несколько алгоритмов результат работы которых объедлинятеся в один ответ:
    3.1.Stacking (диверсификаци, разные алгоритмы)

    3.2.Bagging (везде деревья, диверсификация на уровне данных)
    
    3.3.Бустинг (дефакто стандарт построения ансамбля, по мере итераци исправляются ошибки прошлых алгоритмов, можно бустить до бесконечности, но и время стремится до бесконечности)
    

4. Нейронные сети - каждая единица - это математическая модель

график плотности, который похож на сглаженную гистограмму, группировка данных, 
диаграммы рассеяния)
1. Выявление наиболее важных переменных (должны быть однозначны, не дублировать,
гибкость)
2. Обнаружение выбросов и аномалий
3. Проверка гипотез связанных с конкретной моделью
4. оценка статистик (среднее значение, разброс относительно чего-то) 
и выявление закономерностей
